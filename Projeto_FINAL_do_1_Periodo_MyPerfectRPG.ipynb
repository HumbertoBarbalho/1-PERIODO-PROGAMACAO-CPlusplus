```python
%%writefile Personagens.h
#pragma once // não dá pau ao importar a mesma .h
#include <iostream>
#include <string>
#include <vector>

using namespace std;

struct Protagonista
{
  string nome;
  int vida = 100;
  int vidaMaxima = 100;
  int ataque = 15;
  int defesa = 10;
  int defesaEscudo = 20;
  int defesaBase = 10;
  bool bloqueio = 0;
  int potions= 3;
  int ultrapotions = 0;
  bool Cego = false;
  int turnosCego = 0;
  int chanceAcerto = 95;
  int Moedas = 0;
  int chanceCritico = 15;
  int Critico = 3;
  int turnosVeneno = 0;
};

struct Monstro
{
  string nome;
  int vida;
  int ataque;
  int resistencia;
  int valeMoedas;
  int id;
};

Monstro SpawnarMonstro(int MSorteado)
{
  // Criamos uma lista (vetor) com os monstros
  vector<Monstro> listaDeMonstros =
  {
    {"Goblin Saqueador", 80, 35, 0, 25, 1},      // Posição [0] do Vetor
    {"Esqueleto Guerreiro", 50, 40, 0, 15, 2},    // Posição [1] do Vetor
    {"Morcego Gigante", 70, 30, 0, 10, 3},        // Posição [2] do Vetor
    {"Golem de Pedra", 300, 65, 20, 60, 4},       // Posição [3] do Vetor
    {"Lich Necromante", 150, 30, 10, 40, 5},      // Posição [4] do Vetor
    {"Basilisco Venenoso", 180, 30, 15, 40, 6},  // Posição [5] do Vetor
    {"Death Knight", 500, 70, 50, 10000, 7},        // Posição [6] do Vetor
    {"Orshaball, Demonio do Inferno", 600, 60, 50, 10000, 8}, // Posição [7] do Vetor
    {"Wyvern Maligno", 500, 70, 30, 10000, 9}                  // Posição [8] do Vetor
  };

  //Retorna o monstro que está na posição do número sorteado
  return listaDeMonstros[MSorteado];
}

struct ItemLoja
{
  string nome;
  int preco;
  string tipo; // "pocao", "ataque", "defesa"
  int bonus; // Quanto vai dar de atributo do jogador
  bool disponivel = true;
};

vector<ItemLoja> listaDeItens =
    // NOME  / CUSTO NA LOJA / TIPO DO ITEM / BONUS DADO / SE TA DISPONIVEL
{
  {"Poção de Cura (+30 HP)", 10, "pocao", 1, true},                                   // Posição [0] - Poções são infinitas
  {"Poção de UltraCura (+100 HP)", 50, "ultrapocao", 1, true},                        // Posição [1] - Poções são infinitas
  {"Espada de Aço (40 DE ATQ)", 40, "ataque", 40, true},                              // Posição [2]
  {"Espada Suprema, Excalibur (100 DE ATQ)", 100, "ataque", 100, true},        // Posição [3]
  {"Escudo de Ferro (35 DE DEF BLOQUEIO)", 30, "defesaShield", 35, true},             // Posição [4]
  {"Escudo de Aço Negro (60 DE DEF BLOQUEIO)", 70, "defesaShield", 60, true},         // Posição [5]
  {"Armadura de Mitril (40 de DEF BASE)", 60, "defesaArmor", 40, true},               // Posição [6]
  {"Anel do Cagão (+10% CHANCECRITICA)", 30, "critico", 10, true},                    // Posição [7]
  {"Chaveiro do Pé de Coelho (+10% CHANCE CRITICA)", 60, "critico", 10, true}         // Posição [8]
};

void abrirLoja(Protagonista &jogador)
{
  cout << "\n\n======================== LOJA DA MASMORRA ========================\n" << endl;
  cout << "Vendedor Diz: Bem-vindo Aventureiro!" << endl;

  cout << "\nAtributos Atuais de " << jogador.nome << ":\n" << endl;
  cout << "Dano de Ataque: " << jogador.ataque << endl;
  cout << "Chance Critica: " << jogador.chanceCritico << endl;
  cout << "Defesa de Armadura: " << jogador.defesaBase << endl;
  cout << "Defesa Bonus do Escudo: " << jogador.defesaEscudo << endl;
  cout << "Estoque de Poções: " << jogador.potions << endl;
  cout << "Estoque de UltraPoções: " << jogador.ultrapotions << endl;


  jogador.Cego = false;
  jogador.vida = jogador.vidaMaxima; // Coloquei aqui pra Recuperar a Vida Maxima do Jogador na passagem dos Níveis, quano a Loja é Chamada.

  int ficar = 1;
  while (ficar != 0)
  {
    cout << "\nVocê possui " << jogador.Moedas << " Moedas." << endl;
    cout << "Opções de Itens à Venda: \n" << endl;

    // Exibe os itens dinamicamente usando as posições do vetor
    for (int i = 0; i < listaDeItens.size(); i++) if (listaDeItens[i].disponivel) cout << i + 1 << " - " << listaDeItens[i].nome << " | Custa " << listaDeItens[i].preco << " Moedas" << endl;

    cout << "\nDigite 0 para Sair da Loja" << endl;

    cout << "Vendedor Diz: O que quer comprar?: ";

    int escolha;
    cin >> escolha;

    if (escolha == 0) break; // Se o jogador escolher Sair

    int indice = escolha - 1; // Converte a escolha do menu para o índice do vetor

    if (indice >= 0 && indice < listaDeItens.size() && listaDeItens[indice].disponivel) // Verifica se o item está em estoque com base no ajuste do indice
    {
      ItemLoja &itemSelecionado = listaDeItens[indice]; // Criamos uma referência para o item

      if (jogador.Moedas >= itemSelecionado.preco) //Verifica se o jogador tem dinheiro
      {
        jogador.Moedas -= itemSelecionado.preco;

        // Aplica o efeito baseado no tipo do item Escolhido

        if (itemSelecionado.tipo == "pocao")
        {
          jogador.potions += itemSelecionado.bonus;
          cout << "\nVocê comprou uma poção!" << endl;
          //Aqui não tem mais nada porque poção é infinita
        }

        if (itemSelecionado.tipo == "ultrapocao")
        {
          jogador.ultrapotions += itemSelecionado.bonus;
          cout << "\nVocê Comprou uma Ultra Poção!" << endl;
          //Aqui não tem mais nada porque poção é infinita
        }

        else if (itemSelecionado.tipo == "ataque")
        {
          jogador.ataque = itemSelecionado.bonus;
          itemSelecionado.disponivel = false; // Item único some!
          cout << "\nVocê Comprou a Espada! Ataque aumentado." << endl;
          if (indice == 3) cout << "Você Sente O Poder Verdadeiro da Lamina Lendária do Rei Arthur e sente-se Inspirado!" << endl;
        }
        else if (itemSelecionado.tipo == "defesaArmor")
        {
          jogador.defesaBase = itemSelecionado.bonus; // Usa sua mecânica de escudo
          itemSelecionado.disponivel = false; // Item único some!
          cout << "\nVocê Comprou a Armadura! Defesa Base da Armadura Aumentada." << endl;
        }
        else if (itemSelecionado.tipo == "defesaShield")
        {
          jogador.defesaEscudo = itemSelecionado.bonus; // Usa sua mecânica de escudo
          jogador.bloqueio = true;
          itemSelecionado.disponivel = false; // Item único some!
          cout << "\nVocê Comprou o Escudo! Defesa de Bloqueio Aumentada." << endl;
          if (indice == 5) cout << "Você Percebe a Durabilidade do Material e Sente que pode Confiar sua Vida a Este Escudo" << endl;
        }
        else if (itemSelecionado.tipo == "critico")
        {
          jogador.chanceCritico += itemSelecionado.bonus;
          itemSelecionado.disponivel = false;
          cout << "\nVocê Comprou o Anel! Chance de crítico aumentada." << endl;
          cout << "\nVoce se Sente mais Cagão! (No Bom Sentido)" << endl;
        }
      }

      else cout << "\nVendedor diz: Moedas insuficientes!" << endl;
    }
    else
    {
      cout << "\nOpção inválida ou item esgotado!" << endl;
    }

    cout << "\nAtributos Atuais de " << jogador.nome << ":\n" << endl;
    cout << "Dano de Ataque: " << jogador.ataque << endl;
    cout << "Chance Critica: " << jogador.chanceCritico << endl;
    cout << "Defesa de Armadura: " << jogador.defesaBase << endl;
    cout << "Defesa Bonus do Escudo: " << jogador.defesaEscudo << endl;
    cout << "Estoque de Poções: " << jogador.potions << endl;
    cout << "Estoque de UltraPoções: " << jogador.ultrapotions << endl;

    cout << "\nVendedor Pergunta: Deseja comprar mais Alguma Coisa?\n1 - Sim\n0 - Não\n";
    cin >> ficar;
  }
  cout << "\nSaindo da Loja" << endl;
  cout << "========================================================" << endl;
}
```

```python
%%writefile Combate.h
#pragma once
#include <iostream>
#include <cstdlib>
#include <ctime>
#include "Personagens.h"

using namespace std;

int servos = 0; //Servos do lich declarados de forma Global
bool orshabaalEmFuria = false; // Variável Global pro Orshaball
//LOOP DE COMBATE DO MONSTRO
void turnoDoMonstro(Monstro &inimigo, Protagonista &jogador, int turnoAtual)
{
  cout << "\n**********************TURNO DO MONSTRO**********************" << endl;

  // CALCULA O DANO BASE DO MONSTRO SORTEADO E REGULA BASEADO NA DEFESA
  int danoPadrao = inimigo.ataque - jogador.defesa;
  if (danoPadrao < 1) danoPadrao = 0;

  switch (inimigo.id)
  {
    case 1: // GOBLIN
    if (inimigo.vida < 80 && (rand() % 100) < 50 && turnoAtual % 2 == 0)
    {
      cout << "O Goblin Joga um Dardo de Cegueira na sua fuça!" << endl;
      cout << "Você recebe 5 de dano do Dardo" << endl;
      jogador.turnosCego = 3;
      jogador.Cego = true;
      jogador.chanceAcerto = 35;
    }
    else if (turnoAtual % 2 == 0) cout << "O Goblin Vai Atacar Novamente!"  << endl;
    else
    {
      cout << "O Goblin te causa " << danoPadrao << " de dano!" << endl;
      jogador.vida -= danoPadrao;
    }
    break;

    case 2: // ESQUELETO
    if (turnoAtual % 3 == 0)
    {
      int danoForte = (inimigo.ataque * 2) - jogador.defesa;
      if (danoForte < 1) danoForte = 0; // Evita curar o jogador se a defesa for alta

      cout << "O Esqueleto desfere um GOLPE PODEROSO de " << danoForte << " de dano!" << endl;
      jogador.vida -= danoForte;
      cout << "O Esqueleto Está Cansado e imóvel" << endl;
    }

    else if (turnoAtual % 4 == 0) cout << "O Esqueleto está Cansado e não fez Nada" << endl;
    else if ((turnoAtual + 1) % 3 == 0) // Avisa no turno anterior a pohada forte
    {
      cout << "O Esqueleto Entra em Posição pra ir com Tudo no próximo turno!" << endl;
    }
    else
    {
      cout << "O Esqueleto te ataca e causa " << danoPadrao << " de dano!" << endl; // Se nenhum dos casos anteriores, dano padrão.
      jogador.vida -= danoPadrao;
    }
    break;

    case 3: // MORCEGO GIGANTE

    if (turnoAtual % 2 == 0) cout << "O Morcego voa até você para tentar lhe Morder!" << endl;

    else if (inimigo.ataque > jogador.defesa)
    {
      cout << "O Morcego Gigante Tenta morder seu pescoco e causa " << danoPadrao << " de dano!" << endl;
      cout << "O monstro suga seu sangue e recupera " << danoPadrao << " de vida!" << endl;
      jogador.vida -= danoPadrao;
      inimigo.vida += (danoPadrao);
    }
    else if (inimigo.ataque <= jogador.defesa) cout << "O Monstro Tenta lhe morder mas você o Bloqueia!" << endl;

    break;

    case 4: // GOLEM DE PEDRA

    if (turnoAtual == 1) cout << "O Golem Avisa: NÃO TENHO INTERESSE EM LHE MATAR, HUMANO. MAS SE INSISTIR EM AVANÇAR, AQUI SERÁ SUA COVA!" << endl;
    if (turnoAtual % 3 == 0)
    {
      cout << "o Golem de Pedra lhe acerta um Golpe Massivo de " << danoPadrao << " de Dano" << endl;
      jogador.vida -= danoPadrao;
    }
    else if ((turnoAtual + 1) % 3 == 0) cout << "O Golem lhe Alcança e Levanda os Braços" << endl;

    else cout << "O Golem é Lento para se mover e não lhe Acerta!";
    break;

    case 5: // LICH NECROMANTE

    if (turnoAtual % 2 == 0)
    {
      cout << "O lich Invoca um Zumbi Servo para o Auxiliar em Combate" << endl;
      servos++;
    }

    else
    {
      cout << "O Lich lhe acerta com um raio de Magia Negra e lhe dá "<< danoPadrao << " de Dano!" << endl;
      if (servos == 1) cout << "O Servo do Lich lhe ataca e causa 10 de Dano!" << endl;
      if (servos > 1) cout << "Os " << servos << " Servos do Lich atacam e lhe causam " << (servos*10) << " de Dano!" << endl;
      jogador.vida -= danoPadrao + (10*servos);
    }
    break;

    case 6: // BASILISCO VENENOSO

    if (turnoAtual % 3 == 0) cout << "O Basilisco está Cansado e não faz nada" << endl;
    else
    {
      cout << "O Basilisco Venenoso lhe ataca causando " << danoPadrao << " de dano!" << endl;
      jogador.vida -= danoPadrao;
      if (danoPadrao > 0) jogador.turnosVeneno++; // Adiciona mais 1 nível de veneno que acumula todo turno!
      if ((turnoAtual + 1) % 3 == 0) cout << "O Basilisco se Cansou." << endl;
      cout << "Nível de Veneno Atual: " << jogador.turnosVeneno << endl;
    }

    break;

    case 7: // DEATH KNIGHT

    if (turnoAtual == 1)
    {
      cout << "MORTAL, SOU O GUARDIÃO DESTE ANDAR! MOSTRE SER UM OPONENTE DIGNO!" << endl;
      cout << "O Death Knight avança lentamente desembainhando sua lâmina..." << endl;
    }
    else if (turnoAtual >= 10)
    {
      cout << "\nO Death Knight diz: SEU TEMPO ESGOTOU, MORTAL!" << endl;
      cout << "Ele ergue sua Lâmina Profana e Lhe divide em dois pedaços com um Único Golpe" << endl;
      jogador.vida = 0; // Morte Instantânea
    }
    else
    {
      cout << "O Death Knight desfere um corte pesado com sua Espada Amaldiçoada causando " << danoPadrao << " de dano!" << endl;
      cout << "VOCÊ TEM " << (10 - turnoAtual) << " CHANCES DE ME PROVAR O SEU VALOR!" << endl;
      jogador.vida -= danoPadrao;
    }
    break;

    case 8: // ORSHABALL, DEMONIO DO INFERNO

    if (inimigo.vida < 600)
    {
      int regenerar = 30;
      inimigo.vida += regenerar;
      if (inimigo.vida > 600) inimigo.vida = 600;
      cout << "\nO Orshabaal Ruge e suas Feridas Fecham! Ele recuperou " << regenerar << " de vida." << endl;
    }

    if (inimigo.vida <= 300) // Modo Berserk se ativa se a vida dele cair abaixo de 300
    {
      if (!orshabaalEmFuria) // Se ele AINDA NÃO estava em fúria, exibe o texto
      {
        cout << "ACHA QUE PODE ME FERIR?! MINHAS CHAMAS VÃO DERRETER SUA CARNE!\nSINTA O FOGO DO INFERNO!" << endl;
        cout << "\nO Demônio entra em estado Berserk! Suas garras brilham em chamas intensas!" << endl;
        orshabaalEmFuria = true; // Ativa o booleano para só printar essa frase uma vez
      }
      danoPadrao += 25; // O bônus de +20 continua aplicando normalmente todo turno
    }

    if (turnoAtual == 1) cout << "Adore-me Mortal, Estás na Presença de ORSHABALL!\nLOUVE O MEU PODER ANTES DE MORRER!" << endl;

    else if (turnoAtual % 4 == 0) // Especial do Orshaball
    {
      int danoFogo = danoPadrao + 25;
      cout << "NENHUM MORTAL IRÁ SOBREVIVER!, QUE O CÉU DESABE EM LAVA SOBRE VOCÊ!" << endl;
      cout << "ORSHABAAL INVOCA UMA CHUVA DE METEOROS!" << endl;
      cout << "Você é atingido pelos meteoros e sofre " << danoFogo << " de dano de fogo devastador!" << endl;
      jogador.vida -= danoFogo;
    }
    else if ((turnoAtual + 1) % 4 == 0)
    {
      cout << "Orshabaal Ergue os Braços e Portais de Lava começam a se abrir no Teto da Masmorra!" << endl;
    }
    else
    {
      cout << "Orshabaal avança e te desfere um golpe brutal causando " << danoPadrao << " de dano!" << endl;
      jogador.vida -= danoPadrao;
    }
    break;

    case 9: // WYVERN MALIGNO

    if (turnoAtual == 1) // turno 1
    {
      cout << "UM URRO ENSURDECEDOR ECOA! Um O Wyvern Maligno bate suas Asas Surgindo da Escuridão!" << endl;
      cout << "O Wyvern Bate as asas e Decola, se torna impossivel acerta-lo" << endl;
      jogador.chanceAcerto = 0;
    }

    else if (turnoAtual % 3 == 2) //turnos 2,5,8,11...
    {
      int danoImpacto = danoPadrao + 35;
      cout << "\nO Wyvern Maligno Despenca dos Céus em Velocidade Supersônica!" << endl;
      cout << "O Impacto Esmaga seus Ossos e Causa " << danoImpacto << " de Dano!" << endl;

      jogador.vida -= danoImpacto;
      jogador.chanceAcerto = 95;
      cout << "O Monstro Pousa Pesadamente no Solo, Bufando Fumaça." << endl;
    }
    else if (turnoAtual % 3 == 0) //turnos divisiveis por 3
    {
      cout << "\nO Wyvern Maligno Aproveita que Está no Chão e te Ataca com Garras Afiadas Causando " << danoPadrao << " de dano!" << endl;
      jogador.vida -= danoPadrao;
    }
    else if (turnoAtual % 3 == 1) // esse bloco só roda nos turnos 4, 7, 10...
    {
      cout << "\n[VOO] O Wyvern Maligno dá um Forte Impulso, Bate as Asas e Decola Novamente para o Céu!" << endl;
      cout << "Ele Some Entre as Nuvens Preparando o Próximo Rasante!" << endl;
      jogador.chanceAcerto = 0;
    }
    break;

  }

  cout << "\n************************************************************" << endl;
}

// LOOP DE COMBATE DO JOGADOR
void iniciarCombate(Protagonista &jogador, Monstro &inimigo)
{
  cout << "\n============================================================" << endl;
  int turno = 1;

  //Devem ser Feitos aqui pois é antes da função turno do monstro
  servos = 0; // Reset de mecanica do Lich
  orshabaalEmFuria = false; //Reset de mecanica do Orshaball
  jogador.chanceAcerto = 95;//Reset de Acerto do Wyvern

  while (jogador.vida > 0 && inimigo.vida > 0)
  {
    jogador.bloqueio = false; //Reseta o bloqueio no inicio de cada turno
    jogador.defesa = jogador.defesaBase; // Reseta a defesa ganha pelo escudo

    if (jogador.Cego)
    {
      jogador.turnosCego--;

      if (jogador.turnosCego <= 0)
      {
        jogador.Cego = false;
        jogador.chanceAcerto = 95;
        cout << "\nA cegueira passou! Sua visão voltou ao normal!" << endl;
      }
      else cout << "\nVocê está cego! Chance de acerto reduzida. (" << jogador.turnosCego << " turnos restantes)" << endl;
    }

    if (jogador.turnosVeneno > 0)
    {
      // O dano escala baseado em quantos turnos de veneno foram acumulados
      int danoDoVeneno = jogador.turnosVeneno * 10;
      jogador.vida -= danoDoVeneno;

      cout << "\nO Veneno corrói suas veias! (Nível " << jogador.turnosVeneno << ")" << endl;
      cout << "Você sofreu " << danoDoVeneno << " de dano do veneno acumulado!" << endl;
      if (jogador.vida <= 0)
      {
        cout << "Você não aguenta o Veneno e Morre" << endl;
        jogador.vida = 0;
      }
    }
    int acao;

    cout << "\n[TURNO " << turno << "]" << endl;
    cout << "Sua Vida: " << jogador.vida << "/" << jogador.vidaMaxima;
    cout << " | Vida do " << inimigo.nome << ": " << inimigo.vida << endl;
    cout << "Poções Disponiveis: " << jogador.potions << endl;
    cout << "Ultra Poções Disponiveis: " << jogador.ultrapotions << endl;
    cout << "Quantidade de Moedas: " << jogador.Moedas << endl;
    cout << "1 - Atacar\n2 - Bloquear \n3 - Usar Poção de Cura (+40 HP)\n4 - Usar Ultra Poção (+100 HP)\nEscolha sua Ação: ";
    cin >> acao;

    int danoJogador = jogador.ataque - inimigo.resistencia;
    switch (acao)
    {
      case 1: // ATACAR
      if ((rand() % 100) < jogador.chanceAcerto) // Verifica se o jogador acertou o golpe
      {
        if (inimigo.id == 7 && (rand() % 100) < 50) cout << "\nO Death Knight Intercepta seu Ataque com sua Espada Amaldiçoada e BLOQUEIA COMPLETAMENTE o seu Ataque!" << endl;
        else
        {
          danoJogador = 0;
          bool foiCritico = (rand() % 100) < jogador.chanceCritico; // Verifica se o golpe foi Crítico

          if (foiCritico)
          {
            int ataqueCritico = jogador.ataque * jogador.Critico; // Aplica o multiplicador no dano
            danoJogador = ataqueCritico - inimigo.resistencia;
            if (danoJogador < 0) danoJogador = 0; // Impede que o dano cure.
            inimigo.vida -= danoJogador;

            cout << "\nCRÍTICO! Você aplica um golpe Perfeito ao " << inimigo.nome << " e causou " << danoJogador << " de dano!" << endl;
          }
          else
          {
            danoJogador = jogador.ataque - inimigo.resistencia;
            if (danoJogador < 0) danoJogador = 0; // Impede que o dano cure.
            cout << "\nVocê atacou o " << inimigo.nome << " e causou " << danoJogador << " de dano!" << endl;
            inimigo.vida -= danoJogador; // Retira o dano Calculado final
          }
        }
      }
      else cout << "\nVocê tentou atacar o " << inimigo.nome << ", mas ERROU o golpe!" << endl; // Se o rand der erro
      break;

      case 2: // BLOQUEAR
      jogador.bloqueio = true;
      cout << "\nVocê ergueu seu escudo e se preparou para o próximo ataque!" << endl;
      jogador.defesa = jogador.defesa + jogador.defesaEscudo;
      break;

      case 3: // USAR POÇÃO
      if (jogador.potions > 0)
      {
        jogador.vida += 40;
        if (jogador.vida > jogador.vidaMaxima) jogador.vida = jogador.vidaMaxima;
        jogador.potions--;
        cout << "\nVocê Bebeu uma Poção! Recuperou 40 de vida." << endl;
      }
      else cout << "\nVocê não Tem Poções! Perdeu a Vez!" << endl;
      break;

      case 4: // USAR ULTRAPOÇÃO
      if (jogador.potions > 0)
      {
        jogador.vida = jogador.vidaMaxima;
        cout << "\nVocê Bebeu uma Ultra Poção! Seus efeitos Milagrosos Regeneram sua Vida COMPLETAMENTE!" << endl;
        jogador.ultrapotions--;
      }
      else cout << "\nVocê não Tem Ultra Poções! Perdeu a Vez!" << endl;
      break;

      default: // OPÇÃO INVÁLIDA
      cout << "\nOpção inválida! Você Hesitou e Perdeu a Vez!" << endl;
      break;
    }

    // Turno do Inimigo acontece se ele ainda estiver vivo
    if (inimigo.vida > 0)
    {
    cout << "\n============================================================" << endl;
    turnoDoMonstro(inimigo, jogador, turno);
    }
    turno++;
  }

  if (jogador.vida > 0)
  {
    cout << "\nVoce derrotou o " << inimigo.nome << "! Vitoria!" << endl;
    cout << "\nVocê ganhou " << inimigo.valeMoedas  << " Moedas do Monstro Derrotado!" << endl;
    jogador.Moedas += inimigo.valeMoedas;
    jogador.turnosVeneno = 0;
  }

  else cout << "\nVoce foi derrotado por " << inimigo.nome << "... Fim de Jogo!" << endl;
}
```
